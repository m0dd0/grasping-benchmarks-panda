```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd

# The contact graspnet repo is not packaged and uses relative imports, so we need to
# add it to the path manually.
CONTACT_GRASPNET_PATH = Path("/home/contact_graspnet")
sys.path.append(str(CONTACT_GRASPNET_PATH / "contact_graspnet"))

from data import load_available_input_data
```

```python
TEST_DATA_PATH = CONTACT_GRASPNET_PATH / "test_data"
```

```python
testdata_metadata = {}
for path in sorted(TEST_DATA_PATH.glob("*.npy"), key=lambda x: int(x.stem)):
    segmap, rgb, depth, cam_K, pc_full, pc_colors = load_available_input_data(str(path), K=None)

    testdata_metadata[path.stem] = {
        "segmap_shape": segmap.shape,
        "segmap_dtype": segmap.dtype,
        "segmap_min": np.min(segmap),
        "segmap_max": np.max(segmap),
        "rgb_shape": rgb.shape,
        "rgb_dtype": rgb.dtype,
        "rgb_min": np.min(rgb).tolist(),
        "rgb_max": np.max(rgb).tolist(),
        "depth_shape": depth.shape,
        "depth_dtype": depth.dtype,
        "depth_min": np.min(depth).tolist(),
        "depth_max": np.max(depth).tolist(),
        "cam_K_shape": cam_K.shape,
        "cam_K_dtype": cam_K.dtype,
        "pc_full_shape": pc_full.shape if pc_full is not None else None,
        "pc_full_dtype": pc_full.dtype if pc_full is not None else None,
        "pc_colors_shape": pc_colors.shape if pc_colors is not None else None,
        "pc_colors_dtype": pc_colors.dtype if pc_colors is not None else None,
    }

df = pd.DataFrame.from_dict(testdata_metadata)
df
```
